```python
#Import the libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

#Create a dataframe of the data
data = pd.read_csv('swim_data.csv')
print(data)

#Remove the last row
data = data[:-1]
print(data)

#Create a linear regression model
model = LinearRegression()

#Tell the model what the independent variables are
X = data[['Z0','Z1','Z2','Z3','Z4','Z5']]

#Tell the model what the dependent variable is
y = data['100m Time']

#Fit the model
model.fit(X, y)

#Print the results
print(f"Coefs: {model.coef_}")
print(f"Intercept: {model.intercept_}")

#Make a prediction for the new block of 32,47,11,9,2,4 hours
print(f"Prediction: {model.predict([[32,47,11,9,2,4]])}")
```
